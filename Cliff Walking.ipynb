# <CENTER>CLIFF WALKING

### <I> IMPORTING LIBRARIES

```python
import gym
import numpy as np
import random
import time
from IPython.display import clear_output
import cv2
```

### <I> CREATING ENVIRONMENT

```python
env=gym.make('CliffWalking-v0',render_mode='rgb_array')
```

### <i> Q TABLE

```python
action_space_size=env.action_space.n
state_space_size=env.observation_space.n

print('No of actions: ',action_space_size,'\nNo of states: ',state_space_size)
```

```python
q_table=np.zeros((state_space_size,action_space_size))
print(q_table)
```

### <I>TRAINING

```python
num_episodes=10000
max_steps_per_episode=300
learning_rate=0.1
discount_rate=0.99
exploration_rate=1
max_exploration_rate=1
min_exploration_rate=0.01
exploration_decay_rate=0.001
```

##### 1) Q-LEARNING

```python
rewards_all_episodes=[]


for episode in range(num_episodes):
    
    state=env.reset()[0]
    terminated=False
    rewards_current_episode=0
    
    for step in range(max_steps_per_episode):
        
        exploration_rate_threshold=random.random()
        if exploration_rate_threshold>exploration_rate:
            action=np.argmax(q_table[state,:])
        else:
            action=env.action_space.sample() 
            
        new_state,reward,terminated,truncated,info=env.step(action)
        
        #Updating Q-Table
        q_table[state][action]=q_table[state][action]*(1-learning_rate)+learning_rate*(reward+discount_rate*np.max(q_table[new_state]))
        
        
        state=new_state
        rewards_current_episode+=reward
        
        if terminated or truncated:
            break
    
    # Updating Exploration rate
    exploration_rate=min_exploration_rate+(max_exploration_rate-min_exploration_rate)*np.exp(-exploration_decay_rate*episode)
    
    rewards_all_episodes.append(rewards_current_episode)
    
    
rewards_per_1000_episodes=np.split(np.array(rewards_all_episodes),num_episodes/1000)
count=1000
print('AVERAGE PER 1000 EPISODES')
for r in rewards_per_1000_episodes:
    print('\n',count,' : ',str(sum(r/1000)))
    count+=1000
```

##### 2) SARSA

```python
rewards_all_episodes=[]

def choose_action(state):
    exploration_rate_threshold=random.random()
    if exploration_rate_threshold>exploration_rate:
        action=np.argmax(q_table[state,:])
    else:
        action=env.action_space.sample()
    return action


for episode in range(num_episodes):
    
    state=env.reset()[0]
    terminated=False
    rewards_current_episode=0
    
    action=choose_action(state) 
    
    for step in range(max_steps_per_episode):
            
        new_state,reward,terminated,truncated,info=env.step(action)
        
        new_action=choose_action(new_state) 
        
        #Updating Q-Table
        q_table[state][action]=q_table[state][action]*(1-learning_rate)+learning_rate*(reward+discount_rate*q_table[new_state][new_action])
        
        
        state=new_state
        action=new_action
        rewards_current_episode+=reward
        
        if terminated or truncated:
            break
    
    # Updating Exploration rate
    exploration_rate=min_exploration_rate+(max_exploration_rate-min_exploration_rate)*np.exp(-exploration_decay_rate*episode)
    
    rewards_all_episodes.append(rewards_current_episode)
    
    
rewards_per_1000_episodes=np.split(np.array(rewards_all_episodes),num_episodes/1000)
count=1000
print('AVERAGE PER 1000 EPISODES')
for r in rewards_per_1000_episodes:
    print('\n',count,' : ',str(sum(r/1000)))
    count+=1000
```

### <I>VISUALIZATION

```python
for episodes in range(3):
    state=env.reset()[0]
    terminated=False
    print('EPISODE: ',episodes+1,'\n')
    time.sleep(1)
    for step in range(max_steps_per_episode):

        img=env.render()
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
        cv2.imshow('CliffWalking',img_rgb)
        if cv2.waitKey(25) & 0xFF == ord('q'): 
            break
        time.sleep(0.3)

        action=np.argmax(q_table[state])
        new_state,reward,terminated,truncated,info=env.step(action)
        
        if terminated or truncated:
            img=env.render()
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
            cv2.imshow('CliffWalking',img_rgb)
            if cv2.waitKey(25) & 0xFF == ord('q'): 
                break
                
            if reward==-100:
                print('You fell off the cliff!!')
                time.sleep(3)
                
            else:
                print('You reached the goal!!!')
                time.sleep(3)
                
            
            clear_output(wait=True)
            break
        
        
        state=new_state


cv2.destroyAllWindows()         
env.close()
```
